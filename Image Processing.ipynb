### Concat

```python
fav_1 = pd.concat([aham_data,dhoni_data])
```

### Shuffle

```python
from sklearn.utils import shuffle 
```

```python
my_fav = shuffle(fav).reset_index()
```

```python
my_fav
```

### Index Removal

```python
dream = my_fav.drop(['index'], axis = 1)
```

```python
dream
```

### ML 

```python
x = dream.values[:,:-1]
```

```python
y = dream.values[:,-1]
```

```python
x
```

```python
y
```

### Test and Train Dataset

```python
 from sklearn.model_selection import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, train_size =0.75, test_size = 0.25, random_state = 5)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

### Support Vector Machine

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
```

### SVM Image Prediction

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```
